```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.cluster import DBSCAN
```

```python
data = pd.DataFrame([[3,5],[3,4],[4,4],[9,6],[8,7],[5,10],[8.5,6],[2,3],[5.2,6],[2.5,4]])
```

```python
data
```

```python
plt.scatter(data[0],data[1])
```

```python
dbscan =DBSCAN(eps=2,min_samples=3)
dbscan.fit(data)
```

```python
dbscan.labels_
```

```python
plt.scatter(data[0],data[1],c=dbscan.labels_)
plt.show()
```
